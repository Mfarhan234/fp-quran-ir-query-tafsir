```python
import pandas as pd
import numpy as np

pd.set_option("display.max_colwidth", 200)
```

```python
path = "../data/raw/Al_Quran_Dataset.csv"  
df = pd.read_csv(path)

df.head()
df.info()
```

```python
df.columns
```

```python
cols = [
    "surah_latin",
    "ayah",
    "arabic",
    "translation",
    "tafsir_wajiz",
]

df_use = (
    df[cols]
    .rename(columns={
        "surah_latin": "surah",
        "arabic": "arabic_text",
        "translation": "indonesian_translation",
        "tafsir_wajiz": "tafsir",
    })
    .dropna(subset=["tafsir"])
    .reset_index(drop=True)
)

df_use.head()
```

```python
# misal ID unik: "surah:ayat" + index
df_use["tafsir_id"] = (
    df_use["surah"].astype(str) + ":" + df_use["ayah"].astype(str)
)
df_use.head()
```

```python
out_path = "../data/processed/tafsir_clean.csv"
df_use.to_csv(out_path, index=False)
out_path

```
