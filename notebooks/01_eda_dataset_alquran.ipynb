# Panjang teks tafsir & terjemahan

```python
TEXT_TAFSIR_COL = "tafsir_wajiz"
TEXT_TRANS_COL = "surah_translation"  

df["len_tafsir"] = df[TEXT_TAFSIR_COL].astype(str).str.len()
df["len_translation"] = df[TEXT_TRANS_COL].astype(str).str.len()

df[["len_tafsir", "len_translation"]].describe()
```

# Lihat outlier panjang teks

```python
# Tafsir terpendek
df.sort_values("len_tafsir").loc[:, ["surah_id", "surah_latin", "ayah", TEXT_TAFSIR_COL, "len_tafsir"]].head(10)
```

```python
# Tafsir terpanjang
df.sort_values("len_tafsir", ascending=False).loc[:, ["surah_id", "surah_latin", "ayah", TEXT_TAFSIR_COL, "len_tafsir"]].head(10)
```

# Cek karakter aneh / noise kasar

```python
# Contoh 100 baris acak untuk inspeksi manual
sample_rows = df.sample(100, random_state=42)
sample_rows[[TEXT_TAFSIR_COL]].head(20)
```

```python
# Cek keberadaan pola HTML/basic noise
mask_html = df[TEXT_TAFSIR_COL].astype(str).str.contains("<br>|</p>|&amp;|&quot;", regex=True)
df[mask_html].head(10)
```

# Bentuk DataFrame kerja df_use

```python
cols_use = ["surah_id","surah_latin","ayah","arabic",TEXT_TRANS_COL, TEXT_TAFSIR_COL,]

df_use = df[cols_use].copy()

# Buang baris tanpa tafsir
df_use = df_use.dropna(subset=[TEXT_TAFSIR_COL]).reset_index(drop=True)

df_use.head()
df_use.info()
```

# Tambah tafsir_id dan rename kolom agar konsisten

```python
df_use = df_use.rename(columns={
    "surah_id": "surah_id",
    "surah_latin": "surah",
    "ayah": "ayah",
    "arabic": "arabic_text",
    TEXT_TRANS_COL: "indonesian_translation",
    TEXT_TAFSIR_COL: "tafsir",
})

df_use["tafsir_id"] = df_use["surah"].astype(str) + ":" + df_use["ayah"].astype(str)

df_use.head()
df_use.info()
```

```python
df_use["tafsir_id"].nunique(), len(df_use)
```

# Simpan ke data/processed/tafsir_clean.csv

```python
output_path = "../data/processed/tafsir_clean.csv"
df_use.to_csv(output_path, index=False)
print("Saved to:", output_path)
```
