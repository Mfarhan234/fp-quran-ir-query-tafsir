```python
from pathlib import Path
import sys
import pandas as pd

# kalau notebook ada di /notebooks
ROOT_DIR = Path.cwd().parent.resolve()
print("ROOT_DIR:", ROOT_DIR)

if str(ROOT_DIR) not in sys.path:
    sys.path.append(str(ROOT_DIR))

# load korpus tafsir
PATH_TAFSIR = ROOT_DIR / "data" / "processed" / "tafsir_clean.csv"
df_tafsir = pd.read_csv(PATH_TAFSIR)
print(df_tafsir.shape)
df_tafsir.head()
```

```python
from src.models.relevance_model_loader import (
    AVAILABLE_MODELS,
    load_model_by_name,
    load_best_model,
    score_query_against_corpus,
)

MODELS_DIR = ROOT_DIR / "models"
print("MODELS_DIR:", MODELS_DIR)
print("AVAILABLE_MODELS:", AVAILABLE_MODELS)
```

```python
MODEL_NAME = "XGBoost"   # ganti 'LogisticRegression', 'SVM'

model, scaler, feature_names, use_scaled = load_model_by_name(MODEL_NAME)

print("Model:", type(model))
print("Scaler:", type(scaler) if scaler is not None else None)
print("n_features:", len(feature_names))
print("use_scaled:", use_scaled)
```

```python
best_model, best_scaler, best_feature_names, best_use_scaled, best_name = load_best_model()
print("Best model name:", best_name)
```

```python
import time

user_query = "Saya sering lalai shalat karena kerja, bagaimana pandangan Al-Qur'an soal ini?"

start = time.time()

scores = score_query_against_corpus(
    query_text=user_query,
    df_corpus=df_tafsir,
    model=model,
    scaler=scaler,
    feature_names=feature_names,
    use_scaled=use_scaled,
)

elapsed = time.time() - start
print(f"Waktu total: {elapsed:.2f} detik")
print("Shape skor:", scores.shape)
```
