```python
from pathlib import Path

if (Path("data") / "processed").exists():
    ROOT = Path(".")
else:
    ROOT = Path("..")  # berarti notebook di folder notebooks

PROCESSED_DIR = ROOT / "data" / "processed"

print("ROOT:", ROOT.resolve())
print("PROCESSED_DIR:", PROCESSED_DIR.resolve())
print("Isi processed:", list(PROCESSED_DIR.glob("*")))
```

```python
files = sorted(PROCESSED_DIR.glob("queries_part*.csv"))
files
```

```python
dfs = []
for f in files:
    print("Loading:", f.name)
    df = pd.read_csv(f)
    dfs.append(df)

df_all = pd.concat(dfs, ignore_index=True)
print("Total rows:", len(df_all))
df_all.head()
```

```python
from pathlib import Path

# Sesuaikan ROOT, pakai logika yang tadi
if (Path("data") / "processed").exists():
    ROOT = Path(".")
else:
    ROOT = Path("..")  # kalau notebook ada di folder notebooks

PROCESSED_DIR = ROOT / "data" / "processed"
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

OUTPUT = PROCESSED_DIR / "queries_sintetis_full.csv"
df_all.to_csv(OUTPUT, index=False)

print("Saved merged queries to:", OUTPUT.resolve())
print("Total rows:", len(df_all))
print("Kolom:", df_all.columns.tolist())
```

```python
# pastikan kolom seg_id ada
assert "seg_id" in df_all.columns, "Kolom 'seg_id' tidak ditemukan di df_all!"

unique_ids = set(df_all["seg_id"].unique())
print("Jumlah seg_id unik di queries:", len(unique_ids))

expected_ids = set(range(1, 6237))  # 1 sampai 6236 (6237 eksklusif)

missing = sorted(expected_ids - unique_ids)
extra   = sorted(unique_ids - expected_ids)

print("Ayat yang hilang (seharusnya kosong):", missing)
print("seg_id di luar 1..6236 (seharusnya kosong):", extra)
```
